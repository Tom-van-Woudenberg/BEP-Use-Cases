```python
from sympy.physics.continuum_mechanics.column import Column
import matplotlib.pyplot as plt
```

```python
c = Column(4, 10000, 1)
c.apply_support(0)
c.apply_support(4)
c.apply_telescope_hinge(3)
c.apply_load(50, 2, -1)
c.solve_for_reaction_loads()
```

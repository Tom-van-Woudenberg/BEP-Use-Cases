```python
from sympy.physics.continuum_mechanics.column import Column
```

```python
length, E, A = 12, 40000, 0.5
c = Column(length, E, A)
c.apply_support(0)
c.apply_support(8)
c.apply_support(12)
c.apply_load(10, 3, -1)
c.apply_load(10, 11, -1)
```

```python
c.applied_loads
```

```python
c.load
```

```python
c.solve_for_reaction_loads()
c.reaction_loads
```

```python
c.axial_force()
```

```python
c.deflection()
```
